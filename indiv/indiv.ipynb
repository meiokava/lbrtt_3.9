## Индивидуальное задание
<font size="4"> Вывести количество контуров, найденных с помощью len(контуры).
1. Загружается изображение в оттенках серого
2. Устанавливает пороговое значение для создания двоичного изображения, используя cv2.threshold. Исходное и двоичное изображения затем отображаются с помощью matplotlib.
3. Используется cv2.findContours для поиска контуров в двоичном изображении и cv2.drawContours для рисования контуров на копии исходного изображения. Результирующее изображение с контурами отображается с помощью matplotlib.

```python
import cv2
from matplotlib import pyplot as plt

# Load the image in grayscale
img = cv2.imread('catts.jpg', cv2.IMREAD_GRAYSCALE)

# Threshold the image to create a binary image
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Display the original and binary images
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(122)
plt.imshow(thresh, cmap='gray')
plt.title('Binary')
plt.axis('off')

plt.show()

# Compute the main features of the binary image
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw the contours on a copy of the original image
img_copy = img.copy()
cv2.drawContours(img_copy, contours, -1, (0, 255, 0), 2)

# Display the image with contours
plt.imshow(img_copy, cmap='gray')
plt.title('Contours')
plt.axis('off')
plt.show()

# Print the number of contours found
print(f'Number of contours: {len(contours)}')
```
